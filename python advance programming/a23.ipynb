```python
#2
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        self.head=None
        self.tail=None
        return
    def add_data(self,li):
        for i in li:
            if self.head is None:
                self.head=i
            else:
                self.next=i
            self.tail=i
    def get_data(self):
        out=[]
        current=self.head
        while self.tail!=current:
            out.append(self.val)
            out.append(self.head)
            out.append(self.tail)
            current=self.next
        return out
def add_two_numbers(ll1,ll2):
    output=[]
    l1=ll1.get_data()
    l2=ll2.get_data()
    len_l1=len(l1)
    len_l2=len(l2)
    if(len_l1>len_l2):
        large=l1
        small=l2
    else:
        large=l2
        small=l1
        carry=0
    for i in range(-1,-(len(large)+1),-1):
            if((large[i]+small[i]+carry)<=9):
                output.append(large[i]+small[i]+carry)
            else:
                s=str(large[i]+small[i]+carry)
                output.append(int(s[1]))
                carry=int(s[0])
    output=output[::-1]
    ll3=ListNode(output[0])
    ll3.add_data(output[1:])
    return ll3
lt1 = ListNode(2)
lt1.add_data([4, 3])
lt2 = ListNode(5)
lt2.add_data([6, 4])
print(lt1.get_data()) # [2, 4, 3]
print(lt2.get_data()) # [5, 6, 4]
#print(342 + 465) # 807
print(add_two_numbers(lt1, lt2).get_data())
    
    
    
```

```python
#3
class CoffeeShop:
    def __init__(self,name,menu):
        self.name=name
        self.menu=menu
        self.orders=[]
    def add_order(self,name):
        to_be_added=""
        for item in menu:
            if(item['Name']==name):
                to_be_added=name
        if(to_be_added==""):
            return "This item is currently unavailable!"
        else:
            self.orders.append(name)
            return "Order Added!"
    def fulfill_order(self):
        if(len(self.orders)==0):
            return "All orders have been fufilled!"
        else:
            item=self.orders[0]
            self.orders.remove(item)
            return "The " + item +" is ready!"
    def list_orders(self):
        return self.orders
    def due_amount(self):
        total=0
        for i in self.orders:
            for item in menu:
                if(item['Name']==i):
                    total+=item['Price']
        return total
    def cheapest_item(self):
        price_list=[]
        for item in menu:
            price_list.append(item['Price'])
        mini=min(price_list)
        for item in menu:
            if(item['Price']==mini):
                return item['Name']
    def drinks_only(self):
        drinks=[]
        for item in menu:
            if(item['Type']=="Drink"):
                drinks.append(item['Name'])
        return drinks
    def food_only(self):
        foods=[]
        for item in menu:
            if(item['Type']=="Food"):
                foods.append(item['Name'])
        return foods
menu=[{"Name":"A","Type":"Food","Price":50},{"Name":"B","Type":"Food","Price":70},{"Name":"C","Type":"Drink","Price":20}]
tcs=CoffeeShop("BBQ",menu)
tcs.add_order("D")
tcs.add_order("A")
tcs.add_order("B")
tcs.list_orders()
tcs.due_amount()
tcs.fulfill_order()
tcs.cheapest_item()
tcs.drinks_only()
tcs.food_only()
```

```python
#4

from math import sqrt

def isPrime(N):
    k = int(sqrt(N)) + 1
    for i in range(2, k, 1):
        if (N % i == 0):
            return False

    return True

def getDifference(N,b):
    li=[]
    if (N == 0):
        distance=2
        closest=2
        return (distance,closest)
    elif (N == 1):
        distance=1
        closest=2
        return (distance,closest)

    aboveN = -1
    belowN = -1


    n1 = N + 1
    while (True):
        if (isPrime(n1)):
            aboveN = n1
            break

        else:
            n1 += 1


    n1 = N - 1
    while (True):
        if (isPrime(n1)):
            belowN = n1
            break
            
        else:
            n1 -= 1

    
    diff1 = aboveN - N
    diff2 = N - belowN
    
    if(aboveN>b):
        diff1=diff2
   
    if(diff2==diff1):
        li.append(diff1)
        li.append(belowN)
    elif(diff2>diff1):
        li.append(diff1)
        li.append(aboveN)
    else:
        li.append(diff2)
        li.append(belowN)
    return li
    
    

def loneliest_number(a,b):
    dic={'number':0,'distance':0,'closest':0}
    for i in range(a,b):
        tupl=getDifference(i,b)
        distance=tupl[0]
        closest=tupl[1]
        if(distance>dic['distance']):
            dic['number']=i
            dic['distance']=distance
            dic['closest']=closest
    return dic
          
loneliest_number(0,3)
                
                
    
                
```
